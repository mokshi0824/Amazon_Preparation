```python
def is_palindrome(word):
    cleaned = word.lower()
    return cleaned == cleaned[::-1] and len(cleaned) > 1
def remove_palindromic_words(sentence):
    words = sentence.split()
    result_words = []
    for word in words:
        if not is_palindrome(word):
            result_words.append(word)
    result_sentence = " ".join(result_words)
    return result_sentence
sentence = "In India Malayalam is best with dad and mom"
output = remove_palindromic_words(sentence)
print("Original sentence:", sentence)
print("After removing palindromes:", output)
```
