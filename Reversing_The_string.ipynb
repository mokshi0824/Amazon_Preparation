```python
def reverse_string_keep_symbols(s):
    letters = [ch for ch in s if ch.isalpha()]
    letters.reverse()
    result = []
    j = 0
    for ch in s:
        if ch.isalpha():
            result.append(letters[j])
            j += 1
        else:
            result.append(ch)
    return ''.join(result)
s = "Vision,tranz It :  Solutions"
output = reverse_string_keep_symbols(s)
print("Output:", output)
```
